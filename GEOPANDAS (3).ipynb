```python
kenya_riversproj
nyeri_rivers=gpd.overlay(kenya_riversproj,Nyeri,how="intersection")
nyeri_rivers.plot(linewidth=0.5)
```

```python
import matplotlib.pyplot as plt

# Assuming 'Nyeri' is a DataFrame containing the data you want to plot
fig,ax=plt.subplots()
# Plotting the data
Nyeri.plot(ax=ax,figsize=(5, 5),edgecolor='black',facecolor='none')
nyeri_rivers.plot(ax=ax)
plt.title("Nyeri")
ax.set_xlabel("Eastings")
ax.set_ylabel("Northings")
plt.legend(['County Boundary', 'Rivers'])

# Adjusting the orientation of x-axis labels
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)  # Rotate labels by 45 degrees

plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming 'Nyeri' and 'nyeri_rivers' are DataFrames containing the data you want to plot

# Plotting the data
fig, ax = plt.subplots(figsize=(5, 5))
Nyeri.plot(ax=ax, edgecolor='black', facecolor='none', label='County Boundary')
nyeri_rivers.plot(ax=ax, label='Rivers')
plt.title("Nyeri")
ax.set_xlabel("Eastings")
ax.set_ylabel("Northings")


ax.set_xticklabels(ax.get_xticklabels(), rotation=90)  # Rotate labels by 90 degrees


plt.legend(['County Boundary', 'Rivers'], bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```

```python
import contextily as ctx
import matplotlib.pyplot as plt
from matplotlib.ticker import FixedLocator
import geopandas as gpd

# Ensure the GeoDataFrame is in the correct CRS (EPSG:3857 for web mapping)
Nyeri = Nyeri.to_crs(epsg=3857)

# Plot the GeoDataFrame with a colormap
ax = Nyeri.plot(column="subcounty", figsize=(10, 6), edgecolor='black', legend=True, cmap='viridis')

# Use FixedLocator to set the x-ticks explicitly
xticks = ax.get_xticks()
##ax.xaxis.set_major_locator(FixedLocator(xticks))

# Rotate the x-tick labels
ax.set_xticklabels(xticks, rotation=90)
plt.title("Nyeri")

# Set the axis labels
ax.set_xlabel("EASTINGS")
ax.set_ylabel("NORTHINGS")

# Add a basemap using an alternative provider (e.g., OpenStreetMap)
ctx.add_basemap(ax,crs=Nyeri.crs.to_string(), source=ctx.providers.OpenStreetMap.Mapnik)  # Replace with a valid provider

# Show the plot
plt.show()
```
